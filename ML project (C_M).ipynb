```python
import pandas as pd
import warnings
import matplotlib.pyplot as pt
import numpy as np
%matplotlib inline
warnings.simplefilter("ignore")
```

```python
ga = pd.read_excel("C:\\Users\solit\Downloads\Stock_returns_with_analysis.xlsx")
ga
```

```python
new_data = ga[["ATT", "Nordstrom"]]
new_data
```

```python
x= new_data.iloc[:,0].values.reshape(-1, 1)
y= new_data.iloc[:,1].values.reshape(-1, 1)
print(x)
print(y)
```

```python
pt.scatter(x,y)
pt.xlabel("ATT")
pt.ylabel("Nordstrom")
pt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
comp = pd.DataFrame(x_test, columns = ["ATT"])
comp
comp['Nordstrom"'] = y_test
comp["PREDATA"] = y_predict
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test, y_predict)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
r2_score(y_test, y_predict)
```
